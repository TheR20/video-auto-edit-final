```python
import matplotlib.pyplot as plt
import numpy as np

altura = 169
pixeles_pecho = 90
pixeles_torzo = 278
pecho = (pixeles_pecho/115) * 6 
torso = (pixeles_torzo/115) * 6


def calcular_talla_copa(diferencia_underbust_banda):
  
  tallas_copa = {
      0: "AA",
      1: "A",
      2: "B",
      3: "C",
      4: "D",
      5: "DD",
      6: "E",
      7: "F",
      8: "FF",
      9: "G",
      10: "GG",
  }

  talla_copa = tallas_copa.get(round(diferencia_underbust_banda))

  if talla_copa is None:
    talla_copa = "GG"

  return talla_copa

def calcular_talla_copa_japonesa(diferencia_underbust_banda):
  
  tallas_copa = {
      0: "AA",
      1: "A",
      2: "B",
      3: "C",
      4: "D",
      5: "E",
      6: "F",

  }

  talla_copa = tallas_copa.get(round(diferencia_underbust_banda))

  if talla_copa is None:
    talla_copa = "F"

  return talla_copa

def dibujar_diagrama(pechin,underbustin,finalin):
  # Datos proporcionados
  altura = 185
  pecho = 18.5 
  underbust = 0.432 * altura
  

  # Calculando el radio mayor y menor del óvalo
  radio_mayor = underbust / (2 * np.pi)
  radio_menor = pecho / 2

  # Definiendo las coordenadas de los puntos extremos de las líneas laterales
  x_lateral_izquierda = -radio_menor
  x_lateral_derecha = radio_menor
  y_lateral = -radio_mayor + underbust

  # Definiendo las coordenadas de los puntos extremos de la línea central
  x_centro_izquierda = -radio_menor
  x_centro_derecha = radio_menor
  y_centro = -radio_menor

  # Crear la figura y el eje
  fig, ax = plt.subplots()

  # Dibujar el óvalo que representa el torso
  elipse = plt.Circle((0, 0), radio_menor, color='black', fill=False)
  ax.add_artist(elipse)

  # Dibujar las líneas laterales que representan el pecho
  ax.plot([x_lateral_izquierda, x_lateral_izquierda], [-radio_mayor, y_lateral], color='blue')
  ax.plot([x_lateral_derecha, x_lateral_derecha], [-radio_mayor, y_lateral], color='blue')

  # Agregar texto para mostrar el valor de underbust dentro del óvalo
  ax.text(0, 0, f'Underbust: {underbustin:.2f} cm', horizontalalignment='center', verticalalignment='center')

  # Agregar texto para mostrar el valor de pecho a los lados de las líneas azules
  ax.text(x_lateral_izquierda - 2, radio_mayor - 2, f'Pecho: {pechin:.2f} cm', horizontalalignment='right', verticalalignment='center')
  ax.text(x_lateral_derecha + 2, radio_mayor - 2, f'Pecho: {pechin:.2f} cm', horizontalalignment='left', verticalalignment='center')

  # Definir la posición de la línea central en 3/4 del óvalo
  y_centro = radio_menor * 2/4

  # Dibujar la línea central que atraviesa el óvalo
  ax.plot([x_centro_izquierda, x_centro_derecha], [y_centro, y_centro], color='red')

  # Definir la curva para la parte superior del óvalo
  theta = np.linspace(-np.pi/2, np.pi/2, 100)
  x_curva_izquierda = -radio_menor + radio_menor * np.cos(theta)
  y_curva = np.sqrt((radio_mayor**2) * (1 - ((x_curva_izquierda + radio_menor) / radio_menor)**2))
  x_curva_derecha = radio_menor - radio_menor * np.cos(theta)

  # Dibujar las curvas superiores
  ax.plot(x_curva_izquierda, y_curva+y_centro, color='green')
  ax.plot(x_curva_derecha, y_curva+y_centro, color='green')

  # Agregar texto para mostrar el valor de altura justo arriba de las curvas
  ax.text(0, y_centro + 12, f'Tamaño: {finalin} cm', horizontalalignment='center', verticalalignment='bottom')

  # Agregar el plano cartesiano de fondo
  ax.grid(True, linestyle='--', linewidth=0.5)

  # Configuraciones adicionales
  ax.set_aspect('equal', 'box')
  ax.set_xlim(-radio_mayor * 1.5, radio_mayor * 1.5)
  ax.set_ylim(-radio_mayor * 1.5, radio_mayor * 1.5)
  ax.axis('on')

  # Mostrar el diagrama
  plt.title('Diagrama de Sección Transversal')
  plt.show()

# Calcular el 43.2% de la altura y guardarlo en underbust
underbust = 0.432 * altura
# Multiplicar underbust por pecho por 2.54
resultado = underbust + pecho * 2.54 * 1

diferencia = (resultado - underbust) * 0.3937
talla_copa_americana = calcular_talla_copa(diferencia)
talla_copa_japonesa = calcular_talla_copa_japonesa(diferencia)

# Imprimir el resultado
print("Underbust",underbust)
print("Breast Large",pecho)
print("Torso",torso)
print("Breast Size:", resultado)
print("Difference:", diferencia)
print("American Cup:" , talla_copa_americana)
print("Japanese Cup:" , talla_copa_japonesa)
dibujar_diagrama(pecho,underbust,resultado)
```
